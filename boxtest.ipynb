```python
import functools
import time

from IPython.display import HTML, Image 
import gym
import brax

from brax import envs
from brax.envs import env
from brax import jumpy as jp
from brax.envs import to_torch
from brax.io import html
from brax.io import image
import jax
from jax import numpy as jnp
```

```python
_SYSTEM_CONFIG = """
    bodies {
        name: "skateboard"
        colliders{
            box {
                halfsize { x: 0.45 y: 0.2 z: 0.015}
            }
            material{
                elasticity: 1.0
                friction: 1.0
            }
        }
        inertia{
            x: 1.0
            y: 1.0
            z: 1.0
        }
        mass: 4.00
    }
    bodies {
        name: "floor"
        colliders{
            plane{
            }
            material{
                elasticity: 1.0
                friction: 1.0
            }
        }
        inertia{
            x: 1.0
            y: 1.0
            z: 1.0
        }
        mass: 1.0
        frozen { all: true }
    }
    collide_include{
        first: "floor"
        second: "skateboard"
    }
    defaults{
        qps {
            name: "skateboard"
            pos: {z: 2.0}
            rot: {y: 20}
        }
    }
    friction: 1.0
    gravity {
        z: -9.81
    }
    angular_damping: -0.05
    dt: 0.015
    substeps: 8
    dynamics_mode: "pbd"
"""
```

```python
class TestEnv(envs.Env):
    def __init__(self, config, **kwargs):
        super().__init__(config=config, **kwargs)
    
    def reset(self):
        return self.sys.default_qp()
    def step(self, state, action):
        qp, info = self.sys.step(state.qp, action)
        return qp
    
    def log(self):
        print(self.sys.bodies)

testenv = TestEnv(config=_SYSTEM_CONFIG)

qp = testenv.reset()
# qp.ang[1:5, 1]=1
# qp.vel[0:1, 1] = 20
qps = [qp]
for i in range(100):
#     print(qp)
    print(i)
    qp = testenv.step(env.State(qp = qp, obs=None, reward=None, done=None), jnp.zeros(17))
    qps.append(qp)
# with open("./test.png","wb") as f:
#     f.write(image.render(env.sys, [qp], width=1000, height=1000))

HTML(html.render(testenv.sys, qps))
```
